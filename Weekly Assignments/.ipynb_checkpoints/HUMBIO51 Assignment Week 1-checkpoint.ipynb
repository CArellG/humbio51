```python
#Calculate the length of the Human-BRAF gene and write out the complementary sequence. 
#Note the complementary sequence should be in the 5' to 3' direction. 
```

```python
##ANSWER REMOVE BEFORE GIVING TO STUDENTS
#Read in the sequence and trim the first line
FASTAgenesequence=open('Week_1/data/Human-BRAF-NG_007873.3.txt','r')

genesequence=(FASTAgenesequence.readlines()[1:])

#joins the lines in genesequence into a single string
genesequence=''.join(genesequence)

#removes the linebreaks
genesequence=genesequence.replace('\n','')

#calculates the length of the genesequence
print('The length of the sequence is ' + str(len(genesequence)))

complementarysequence=''

for i in genesequence:
    if i=='A':
        complementarysequence=complementarysequence+'T'
    if i=='T':
        complementarysequence=complementarysequence+'A'
    if i=='G':
        complementarysequence=complementarysequence+'C'
    if i=='C':
        complementarysequence=complementarysequence+'G'
print ('The complementary sequence is :\n'+ complementarysequence[::-1])
```

## Question 4

The 'Cracking your Genetic Code' documentary had the story of a patient using the cystic fibrosis drug Kalydeco.

One of the challenges of treating cystic fibrosis is that different mutations in the same protein can lead to the disease. Kalydeco is effective in patients in which the glycine (or G) at position 551 has been changed to an aspartic acid (or D). 

This is noted as a G551D mutation.

The sequence for the CFTR protein is provided in Week_1/data/Human-CFTR-NP_000483.3.txt. Write a script to read in the protein sequence and change the G at position 551 to a D.  

Have the script also print the length of both sequences. 

```python
##ANSWER REMOVE BEFORE GIVING TO STUDENTS

FASTAproteinsequence=open('Week_1/data/Human-CFTR-NP_000483.3.txt','r')

proteinsequence=(FASTAproteinsequence.readlines()[1:])

#joins the lines in genesequence into a single string
proteinsequence=''.join(proteinsequence)

#removes the linebreaks
proteinsequence=proteinsequence.replace('\n','')

print('The length of the CFTR sequence is ' + str(len(proteinsequence)))

print('The amino acid at position 551 in the CFTR protein is ' + proteinsequence[550])

mutated_proteinsequence=proteinsequence[0:550] + "D" + proteinsequence[551:]


print('The length of the G551D CFTR sequence is ' + str(len(proteinsequence)))

print('The amino acid at position 551 in the G551D CFTR sequence is ' + mutated_proteinsequence[550])

```
